### 1. Which establishments have a hygiene score equal to 20?

```python
# Query to find establishments with a hygiene score greater than or equal to 20
query = {'scores.Hygiene': {'$gte': 20}}

# Use count_documents to count the number of matching documents
results = establishments.count_documents(query)

# Print the count
print(f"There are {results} documents with Hygiene scores greater than or equal to 20")

# Fetch all documents matching the query
establishments20 = list(establishments.find(query))

# Display the first document in the results
firstEstablishment = establishments20[0]

# Display the first establishment
print(f"The first establishment in the result with a hygiene score of more than 20 is: {firstEstablishment.get('BusinessName')}")
```

```python
# Convert the result to a Pandas DataFrame
df = pd.DataFrame(establishments20)

# Display the number of rows in the DataFrame
numRows = len(df)
print(numRows)

# Display the first 10 rows of the DataFrame
df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query = {
    'AddressLine4': { '$regex': '^London$'},  #Regex for London
    'RatingValue': {'$gte': 4}  # RatingValue >= 4
}


# Use count_documents to display the number of documents in the result
results = establishments.count_documents(query)
print(results)

# Display the first document in the results using pprint
# Fetch the first result
result = results[0]

# Pretty-print the result
pprint(result)
```

```python
# Convert the results to a DataFrame
query = query ={'LocalAuthorityName':{'$regex':'London'},'RatingValue':{'$gte': 4}}

queryResults = establishments.find(query)

# Put results to DF
df = pd.DataFrame(list(queryResults))

# Display the number of rows in the DataFrame
rowCount = len(df)
print(f"There are {rowCount} in the DataFrame")

# Display the first 10 rows of the DataFrame
df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Set parameters for lat and long search

# Set parameters for lat and long search
latLong = {
    'geocode.latitude': 51.49014200,
    'geocode.longitude': 0.08384000    
}

# Calculate bounds in Python
lat_min = latLong["geocode.latitude"] - 0.01
lat_max = latLong["geocode.latitude"] + 0.01
long_min = latLong["geocode.longitude"] - 0.01
long_max = latLong["geocode.longitude"] + 0.01

# Build the query
query = {
    'geocode.latitude': {'$gte': lat_min, '$lte': lat_max},
    'geocode.longitude': {'$gte': long_min, '$lte': long_max},
    'RatingValue': {'$eq': 5}
}

# Fetch, sort, and limit results
results = list(
    establishments.find(query)
    .sort('scores.Hygiene', 1)
    .sort('BusinessName', 1)
    .limit(5)
)

# Print the results
pprint(results)
```

```python
# Convert result to Pandas DataFrame
df = pd.DataFrame(list(results))
df.head()
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest

# Set up queries for the pipeline
matchQuery = {'$match':{'scores.Hygiene':{'$eq':0}}}
groupQuery = {'$group': {'_id': '$LocalAuthorityName', 'count': {'$sum': 1}}}
sortQuery = {'$sort':{'count': -1}}

# Execute the pipeline
pipeline = [matchQuery, groupQuery, sortQuery]

# Print the number of documents in the result
results = list(establishments.aggregate(pipeline))
docLength = len(results)
print(f"There are {docLength} documents in the results set.")


# Print the first 10 results
print("Top 10 Local Authorities with establishments having a hygiene score of 0:")
for result in results[:10]:  # Slice the results list to get the first 10 documents
    print(f"Local Authority: {result['_id']}, Count: {result['count']}")
```

```python
# Convert the result to a Pandas DataFrame
df = pd.DataFrame(results)
# Display the number of rows in the DataFrame
rowCount = len(df)
# Display the first 10 rows of the DataFrame
df.head(10)
```
